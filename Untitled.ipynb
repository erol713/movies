```python
import numpy as np
import pandas as pd
import datetime
```

```python
df = pd.read_csv('myWatchedMovies.csv', index_col=0)
```

```python
df.head()
```

```python
df.columns = [ 'Watched', 'Date', 'Day', 'Rating', 'Desc', 'year filmed', 'director', 'genre', 'key words']
```

```python
df.head()
```

```python
df = df[(df['Watched']== 'Me') | (df['Watched']== 'Berina') | (df['Watched']== 'Mob ;Aida;Munevera')]
```

```python
df['Watched'].value_counts()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.sort_values(by='Date')
```

```python
df.loc[df.Watched == 'Me', 'Watched'] = "Alone"
df.loc[df.Watched == 'Berina', 'Watched'] = "with gf"
df.loc[df.Watched == 'Mob ;Aida;Munevera', 'Watched'] = "Family"
```

```python
df['Day'] = df['Date'].dt.weekday
```

```python
df.loc[df.Day == 0, 'Day'] = "monday"
df.loc[df.Day == 1, 'Day'] = "tuesday"
df.loc[df.Day == 2, 'Day'] = "wednesday"
df.loc[df.Day == 3, 'Day'] = "thursday"
df.loc[df.Day == 4, 'Day'] = "friday"
df.loc[df.Day == 5, 'Day'] = "saturday"
df.loc[df.Day == 6, 'Day'] = "sunday"
```
