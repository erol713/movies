```python
import pandas as pd
import numpy as np
import datetime as dt
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data = pd.read_csv('final_data_movies.csv', index_col=0)
mine = pd.read_csv('final_my_movies.csv', index_col=0)
```

```python
data = data.set_index('movieId')
mine = mine.set_index('movieId')
```

```python
family = mine[mine['Watched'] == 'with family']
family.drop(['Watched', 'Day'], axis=1, inplace=True)

alone = mine[mine['Watched'] == 'Alone']
alone.drop(['Watched', 'Day'], axis=1, inplace=True)

gf = mine[mine['Watched'] == 'with gf']
gf.drop(['Watched', 'Day'], axis=1, inplace=True)

workdays = mine[(mine['Watched'] != 'sunday') & (mine['Day'] != 'saturday')]
workdays.drop(['Watched', 'Day'], axis=1, inplace=True)

weekend = mine[(mine['Day'] == 'sunday') | (mine['Day'] == 'saturday')]
weekend.drop(['Watched', 'Day'], axis=1, inplace=True)
```

```python
weekend.drop(['title'], axis=1, inplace=True)
workdays.drop(['title'], axis=1, inplace=True)
alone.drop(['title'], axis=1, inplace=True)
gf.drop(['title'], axis=1, inplace=True)
family.drop(['title'], axis=1, inplace=True)
```

```python
family = family.join(pd.get_dummies(family.keywords))
family.drop('keywords', axis=1, inplace=True)
family = family.join(pd.get_dummies(family.genres))
family.drop('genres', axis=1, inplace=True)

alone = alone.join(pd.get_dummies(alone.keywords))
alone.drop('keywords', axis=1, inplace=True)
alone = alone.join(pd.get_dummies(alone.genres))
alone.drop('genres', axis=1, inplace=True)

gf = gf.join(pd.get_dummies(gf.keywords))
gf.drop('keywords', axis=1, inplace=True)
gf = gf.join(pd.get_dummies(gf.genres))
gf.drop('genres', axis=1, inplace=True)

workdays = workdays.join(pd.get_dummies(workdays.keywords))
workdays.drop('keywords', axis=1, inplace=True)
workdays = workdays.join(pd.get_dummies(workdays.genres))
workdays.drop('genres', axis=1, inplace=True)

weekend = weekend.join(pd.get_dummies(weekend.keywords))
weekend.drop('keywords', axis=1, inplace=True)
weekend = weekend.join(pd.get_dummies(weekend.genres))
weekend.drop('genres', axis=1, inplace=True)
```

```python
alone_weekend = pd.concat([alone, weekend], axis=1)
alone_workdays = pd.concat([alone, workdays], axis=1)

gf_weekend = pd.concat([gf, weekend], axis=1)
gf_workdays = pd.concat([gf, workdays], axis=1)

family_weekend = pd.concat([family, weekend], axis=1)
family_workdays = pd.concat([family, workdays], axis=1)
```

```python
alone_weekend = alone_weekend.dropna()
alone_workdays = alone_workdays.dropna()

gf_weekend = gf_weekend.dropna()
gf_workdays = gf_workdays.dropna()

family_weekend = family_weekend.dropna()
family_workdays = family_workdays.dropna()
```

```python
alone_weekend = alone_weekend.loc[:,~alone_weekend.columns.duplicated()]
alone_weekend = alone_weekend[alone_weekend['vote_average'] > 5.0]

alone_workdays = alone_workdays.loc[:,~alone_workdays.columns.duplicated()]
alone_workdays = alone_workdays[alone_workdays['vote_average'] > 5.0]

gf_weekend = gf_weekend.loc[:,~gf_weekend.columns.duplicated()]
gf_weekend = gf_weekend[gf_weekend['vote_average'] > 5.0]

gf_workdays = gf_workdays.loc[:,~gf_workdays.columns.duplicated()]
gf_workdays = gf_workdays[gf_workdays['vote_average'] > 5.0]

family_weekend = family_weekend.loc[:,~family_weekend.columns.duplicated()]
family_weekend = family_weekend[family_weekend['vote_average'] > 5.0]

family_workdays = family_workdays.loc[:,~family_workdays.columns.duplicated()]
family_workdays = family_workdays[family_workdays['vote_average'] > 5.0]

```

```python
my_ratings = pd.read_csv('myWatchedMovies2.csv', index_col=0)
```

```python
my_ratings = my_ratings[['Watched', 'Day', 'Rating', 'movieId', 'title']]
```

```python
def get_recomm_movies(x):
    Filtering_cosim = cosine_similarity(x,x)
    best_movies = []
    recommendations = []
    
    for ind, i in enumerate(x.index):
        if my_ratings.loc[i].Rating >= 3.0:
            best_movies.append(ind)
    
    for i in best_movies:
        most_sim_movies = sorted(list(enumerate(Filtering_cosim[i])), key=lambda x: x[1], reverse=True)
        most_sim_movies = most_sim_movies[1:11]
        sim_movies = [x[0] for x in most_sim_movies]
        for i in sim_movies:
            if (data.iloc[i].title) in recommendations:
                continue
            else:                
                recommendations.append(data.iloc[i].title)
        
    return recommendations
```
